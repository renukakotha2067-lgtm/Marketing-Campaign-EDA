## 1. Import Libraries

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
sns.set_theme()
import warnings; warnings.filterwarnings('ignore')
```

## 2. Load Dataset

```
df=pd.read_excel('Market Campaign dataset.xlsx')
df.head()
```

## 3. Explore Dataset

```
display(df.shape)
display(df.info())
display(df.describe())
display(df.isnull().sum())
```

## 4. Data Cleaning

```
df['Income']=df['Income'].fillna(df['Income'].median())
df=df.drop_duplicates().reset_index(drop=True)
df['Dt_Customer']=pd.to_datetime(df['Dt_Customer'])
```

## 5. Feature Engineering

```
df['Age']=2026-df['Year_Birth']
df['Total_Spending']=df[['MntWines','MntFruits','MntMeatProducts','MntFishProducts','MntSweetProducts','MntGoldProds']].sum(axis=1)
df['Total_Purchases']=df[['NumWebPurchases','NumCatalogPurchases','NumStorePurchases']].sum(axis=1)
df['Accepted_Campaigns']=df[['AcceptedCmp1','AcceptedCmp2','AcceptedCmp3','AcceptedCmp4','AcceptedCmp5']].sum(axis=1)
```

## 6. EDA

```
fig,axs=plt.subplots(2,2,figsize=(12,8))
sns.histplot(df['Income'],ax=axs[0,0]);axs[0,0].set_title('Income')
sns.barplot(x='Education',y='Total_Spending',data=df,ax=axs[0,1]);axs[0,1].tick_params(axis='x',rotation=30)
df[['NumWebPurchases','NumCatalogPurchases','NumStorePurchases']].sum().plot(kind='bar',ax=axs[1,0],title='Purchases by Channel')
sns.scatterplot(x='Income',y='Total_Spending',data=df,ax=axs[1,1])
plt.tight_layout();plt.show()

plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True),cmap='coolwarm')
plt.show()
```

## 7. Business Insights

- Higher-income customers spend more.
- Wine products contribute the largest spending.
- Store purchases exceed web/catalog purchases.
- Previous campaign responders are valuable targets.
